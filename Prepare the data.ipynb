```python
import os
```

```python
path = ["C:\\Users\\dell\\Desktop\\projects\\Analyzing Movie Reviews Sentiment\\aclImdb\\train\\neg",
        "C:\\Users\\dell\\Desktop\\projects\\Analyzing Movie Reviews Sentiment\\aclImdb\\train\\pos",
        "C:\\Users\\dell\\Desktop\\projects\\Analyzing Movie Reviews Sentiment\\aclImdb\\test\\neg",
        "C:\\Users\\dell\\Desktop\\projects\\Analyzing Movie Reviews Sentiment\\aclImdb\\test\\pos"]
```

```python
ll = []
for p in path:
    #print(p)
    filenames = os.listdir(p)
    ll.append(filenames)
```

```python
dd = ["negative", "positive", "negative", "positive"]
```

```python
v=0
err = []
outfile = open("movie_reviews.txt", "w",encoding="utf8")
outfile1 = open("y.txt", "w",encoding="utf8")
for l in ll:
    p = path[ll.index(l)]
    y = dd[ll.index(l)]
    for i in l:
        try:
            infile = open(p+"\\"+i,mode='r',encoding="utf8")
            contents = infile.read()
            outfile.write("\"" + contents + "\"\n")
            outfile1.write("\"" + y + "\"\n")

        except:
            err.append(i)
            v+=1
outfile.close()
outfile1.close()
```
